```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.txt', sep=';', names=[ 'dataLen', 'threadsCount', 'time'])
```

```python
data.head()
```

```python
data[data['dataLen'] == 10000000]
```

```python
df_10mln = data[data['dataLen'] == 10000000]
df_500mln = data[data['dataLen'] == 500000000]
df_1bln = data[data['dataLen'] == 1000000000]
```

```python
df_10mln
```

```python
plt.plot(df_10mln['threadsCount'], df_10mln['time'], label = '10mln')
plt.title('Зависимость числа потоков от времени для входных данных в 10млн')
plt.xlim(1, 15)
plt.legend()
```

```python
df_500mln
```

```python
plt.plot(df_500mln['threadsCount'], df_500mln['time'], label = '500mln')
plt.title('Зависимость числа потоков от времени для входных данных в 500млн')
plt.xlim(1, 15)
plt.legend()
```

```python
df_1bln
```

```python
plt.plot(df_1bln['threadsCount'], df_1bln['time'], label = '1bln')
plt.title('Зависимость числа потоков от времени для входных данных в 1млрд')
plt.xlim(1, 15)
plt.legend()
```

```python
plt.plot(df_1bln['threadsCount'], df_1bln['time'], label = '1bln')
plt.plot(df_10mln['threadsCount'], df_10mln['time'], label = '10mln')
plt.plot(df_500mln['threadsCount'], df_500mln['time'], label = '500mln')
plt.title('Зависимость числа потоков от времени при различных входных данных')
plt.xlim(1, 15)
plt.ylim(0, 60000)
plt.legend()
```
